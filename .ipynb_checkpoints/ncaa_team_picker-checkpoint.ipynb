```python
import pandas as pd
import duckdb as db
```

```python
df = pd.read_excel('sportsref_download(1).xlsx')
df
```

```python
schools = db.query("SELECT distinct School FROM df where School like '%NCAA%'").to_df()
schools
```

```python
offense = df.sort_values(by=['Tm.'],  ascending=False).reset_index().reset_index()
offense =  offense.rename(columns={'level_0': 'offense'})
offense[['School', 'offense']]
```

```python
defense = df.sort_values(by=['Opp.'],  ascending=True).reset_index().reset_index()
defense =  defense.rename(columns={'level_0': 'defense'})
defense[['School', 'defense', 'Opp.']]
```

```python
SOS = df.sort_values(by=['SOS'],  ascending=False).reset_index().reset_index()
SOS =  SOS.rename(columns={'level_0': 'SOS_2'})
SOS[['School', 'SOS_2']]
```

```python
FT = df.sort_values(by=['FT%'],  ascending=False).reset_index().reset_index()
FT =  FT.rename(columns={'level_0': 'FT_pct'})
FT[['School', 'FT_pct']]
```

```python
FTA = df.sort_values(by=['FTA'],  ascending=False).reset_index().reset_index()
FTA =  FTA.rename(columns={'level_0': 'FT_att'})
FTA[['School', 'FT_att']]
```

```python
ncaa_analysis = db.query('''SELECT s.School, offense+1 offense, defense+1 defense, 
         SOS_2+1 SOS, FT_pct+1 FT_pct, FT_att+1 FT_att FROM schools s
         left join offense o
         on o.School=s.School
         left join defense d
         on d.School=s.School
         left join SOS sos
         on sos.School=s.School
         left join FT fp
         on fp.School=s.School
         left join FTA fa
         on fa.School = s.School''').to_df()
ncaa_analysis
```

```python
ncaa_analysis['FT_ave'] = (ncaa_analysis['FT_pct']+ncaa_analysis['FT_att'])/2
```

```python
df_analysis = pd.read_csv('analysis.csv')
df_analysis
```

```python
ncaa_analysis_2 = db.query('''select a.*, seeding from ncaa_analysis a
                            left join df_analysis d
                            on d.School=a.School''').to_df()
ncaa_analysis_2
```

```python
db.query('''select * from ncaa_analysis_2
            where offense <= 100
            and defense <= 100''').to_df()
```
